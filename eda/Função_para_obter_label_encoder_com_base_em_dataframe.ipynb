#Funcção para obtenção de valores

```python
def getLabelEncoderCategory(df='dataframe',field_name='name'):
    """
    Retorna os valores codificados para uma nova coluna denominada encoder_<field_name> para o campo passado como parametro adicinando-o ao dataframe

    :param df: dataframe
    :param fild_name: string

    :return dataframe:
    """
    #import matplotlib.pyplot as plt
    from sklearn.preprocessing import LabelEncoder
    field_encoder = LabelEncoder()
    le_array_res = field_encoder.fit_transform(df[field_name])
    df['Encoder_'+field_name] = le_array_res
    return df
```

# Carregando dataset de exemplo e aplicando a função

### Dados sem aplicação da função

```python
import pandas as pd
dados = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv",sep=',')
dados.sort_values(by=['sepal_width']).head()
```

## Dados após aplicação da função. Observe que agora temos uma nova coluna com prefixo Encoder que é numérica e representa as espécies

```python
dados = getLabelEncoderCategory(df=dados,field_name='species')
dados.sort_values(by=['sepal_width']).head()
```
