#Conectando ao google drive e listando conteúdo de uma pasta que possui o arquivo de funcoes denominado 'auxiliar_functions.py'

```python
def conecta_drive_lista_conteudo(folder_target='pasta_alvo',show_path='False'):
    """
    Lista os arquivos ou pasta na pasta/diretório do google drive indicados como parametro

    :param folder_target: path
    :param show_path: bool
    :return files: files or folders

    """
    # importacoes
    import os
    #!pip install google-colab
    from google.colab import drive

    #montando google drive
    drive.mount('/content/drive')

    # Define o caminho para o diretório desejado no Google Drive
    caminho_drive = '/content/drive/My Drive/'+folder_target

    # Lista os arquivos no diretório
    files = os.listdir(caminho_drive)
    if show_path =='True':
        return (files,caminho_drive)
    else:
        return (files)
```

```python
res = conecta_drive_lista_conteudo('etec',show_path='True')
res
```

##### Observe que a função retorna tanto os aquivos quanto o caminho (obtido por meio do parametro show_path = 'True')
