```python
def dataframe_to_list_and_drop_none_values(df='dataframe'):
    """
    Retorna dataframe convertido em uma lista excluidos os valores nulos ('None') existentes no dataframe.
    Ou sej, a função possibilita obter uma lista excluindo os valores nulos a partir de um dataframe

    :param df: dataframe

    :return cleaned_list: list

    """
    arr_1 = df.to_numpy()
    list_1 = arr_1.tolist()
    cleaned_list = []
    for sublist in list_1:
        cleaned = [elem for elem in sublist if elem is not None]
        if len(cleaned):  # anything left?
            cleaned_list.append(cleaned)
    #print(f'List: {cleaned_list}')
    return cleaned_list

def analyse_asssociation_rules(df='dataframe',array_fields_assoc=['field','field2'],show_frequent_itens='False'):
    """
    Esta função possibilita analisar as regras de associação no conjunto de dados repassado

    :param data: dataframe

    """
    from mlxtend.preprocessing import TransactionEncoder
    from mlxtend.frequent_patterns import apriori
    from mlxtend.frequent_patterns import association_rules
    te = TransactionEncoder()
    list_from_dataframe = dataframe_to_list_and_drop_none_values(df)
    te_ary = te.fit(list_from_dataframe).transform(list_from_dataframe)
    #print(te_ary)
    te_ary.astype("int")
    #te.columns_
    dfres = pd.DataFrame(te_ary, columns=te.columns_)
    frequent_itemsets = apriori(dfres, min_support=0.4, use_colnames=True)
    if show_frequent_itens=='True':
        print(frequent_itemsets)
    rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.4)
    return rules.sort_values('confidence', ascending=False)
```

```python
import pandas as pd
sales_transaction = pd.DataFrame([['T-shirt', 'Trousers', 'Belt'],
           ['T-shirt', 'Jacket'],
           ['Jacket', 'Gloves'],
           ['T-shirt', 'Trousers', 'Belt'],
           ['T-shirt', 'Trousers', 'Sneakers', 'Jacket', 'Belt'],
           ['Jacket', 'Belt'],
           ['Trousers', 'Sneakers', 'Belt'],
           ['Trousers', 'Belt', 'Sneakers'],
           ['T-shirt', 'Jacket','Trousers', 'Belt']])

analyse_asssociation_rules(df=sales_transaction,show_frequent_itens='False')
```
